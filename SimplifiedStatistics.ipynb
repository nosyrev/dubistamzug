```python
import pandas as pd

# Aggregated statistics for the whole project
map_paths = [
    "Standorte Plakate DBAZ 2024  1. & 2.csv",
    "Standorte DBAZ 2024 Woche 4.csv",
]

map_df = pd.DataFrame()
for path in map_paths:
    df = pd.read_csv(path)
    df["map_name"] = path
    map_df = pd.concat([map_df, df])
map_df.head()
```

```python
print("Point statistcs")

print("Total: ", len(map_df))

for tt, tt_df in map_df.groupby("type"):
    print(f"{tt}: {len(tt_df)}")
```
